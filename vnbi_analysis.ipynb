```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read file
```

```python
data = pd.read_json('post.json')
data.head(5)
```

```python
# Check null
```

```python
pd.isna(data).sum()
```

```python
# Convert type datetime
```

```python
data['date_post'] = pd.to_datetime(data['date_post'])
```

```python
data.dtypes
```

```python
# Thống kê số lượng bài viết theo tháng trong vòng 1 năm
```

```python
data['month'] = data['date_post'].dt.month
data['year'] = data['date_post'].dt.year
```

```python
# Lấy ra các bài post trong vòng 1 năm
```

```python
one_year_ago = pd.to_datetime('today') - pd.DateOffset(years=1)
one_year_ago = pd.to_datetime(one_year_ago.date())
```

```python
data_one_year_ago = data[data['date_post'] >= one_year_ago]
```

```python
data_one_year_ago
```

```python
# Bảng dưới thể hiện số bài được viêt theo từng tháng trong vòng 1 năm
```

```python
total_post_month = data_one_year_ago.groupby(['month', 'year']).size().reset_index(name='quantity')
total_post_month
```

```python
total_post_month['date'] = total_post_month.apply(lambda row: f"{row['year']}-{row['month']}", axis=1)
total_post_month['date'] = pd.to_datetime(total_post_month['date'],format='%Y-%m')
total_post_month = total_post_month.sort_values('date')
total_post_month['date'] = total_post_month['date'].dt.strftime('%Y-%m')
```

```python
# Đồ thị dạng cột thể hiện số bài được viêt theo từng tháng trong vòng 1 năm
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.bar(total_post_month['date'], total_post_month['quantity'])
plt.xlabel("Date")
plt.ylabel("Quantity")
```
